```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=100
x=np.arange(1,n+1)
y=2*x+1+50*(np.random.random([n])-0.5)
ideal_model=2*x+1
np.random.seed(0)
w,b=2*np.random.random(2)
print(f'initial value: w={w}, b={b}')
lr = 1e-5
model=w*x+b
```

```python
train_loss=[]
for i in range(100):
    dL_dw = (2/n) * x @ (model.T-y.T)
    dL_db = (2/n) * np.sum(model-y)
    w -= lr * dL_dw
    b -= lr * dL_db
    model = w*x + b
    L = np.sum((model-y)**2)/n
    train_loss.append(L)
```

```python
plt.scatter(x,y,c='green',label='target data')
plt.plot(x,ideal_model,c='r',label='ideal_model',linewidth=5)
plt.plot(x,model,c='b',label='model',linewidth=5)
xlabel=plt.xlabel('input')
ylabel=plt.ylabel('output')
plt.legend()
plt.show()
plt.plot(np.arange(1,101), train_loss, label='train_loss')
plt.legend()
plt.show()
```
